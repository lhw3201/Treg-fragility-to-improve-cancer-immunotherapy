**Reference Solution**

```python
import numpy as np
from scipy.stats import wasserstein_distance

def solve_scicode():
    """
    Infer the population-level shift in Treg fragility using Earth Mover's Distance.
    """

    # 1. Input Data
    control_data = np.array([
        [2.1, 1.8, 0.1, 0.2, 0.5],
        [1.9, 2.0, 0.0, 0.1, 0.4],
        [2.5, 2.2, 0.2, 0.3, 0.6]
    ])

    treated_data = np.array([
        [0.5, 0.6, 2.8, 2.5, 2.1],
        [1.1, 0.9, 1.5, 1.2, 1.8],
        [0.8, 0.4, 3.1, 2.9, 2.5]
    ])

    # Define indices based on biological roles
    stability_indices = [0, 1]  # Foxp3, Ikzf2
    fragility_indices = [2, 3, 4] # Ifng, Tbx21, Gzmb

    # 2. Normalization Logic
    # We Z-score genes across the global population to ensure comparable scales
    combined_data = np.vstack([control_data, treated_data])
    global_means = np.mean(combined_data, axis=0)
    global_stds = np.std(combined_data, axis=0) + 1e-9 # Handle zero variance

    def calculate_cell_indices(data):
        # Apply Z-score
        z_scores = (data - global_means) / global_stds

        # Calculate Stability Score and Fragility Score per cell
        s_score = np.mean(z_scores[:, stability_indices], axis=1)
        f_score = np.mean(z_scores[:, fragility_indices], axis=1)

        # Fragility Index: Sigmoid of the difference
        # Index > 0.5 suggests a shift toward a fragile phenotype
        diff = f_score - s_score
        return 1 / (1 + np.exp(-diff))

    # 3. Distribution Inference
    control_indices = calculate_cell_indices(control_data)
    treated_indices = calculate_cell_indices(treated_data)

    # 4. Quantification of Shift
    # Using Wasserstein-1 distance to measure the "work" to shift the population
    emd_score = wasserstein_distance(control_indices, treated_indices)

    return float(emd_score)

# Final output check
if __name__ == "__main__":
    print(solve_scicode())
```
